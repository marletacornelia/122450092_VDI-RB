**Visualisasi**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#Membuat dataset
dataLagu = {
    'Usia': [18, 25, 30, 22, 35, 40, 28, 50, 19, 45],
    'Genre Musik': ['Pop', 'Rock', 'Jazz', 'Pop', 'Classical', 'Rock', 'Jazz', 'Classical', 'Pop', 'Classical'],
    'Durasi Waktu (Menit per Hari)': [120, 150, 90, 110, 60, 180, 100, 45, 130, 70]
}

df = pd.DataFrame(dataLagu)

#Membuat plot menggunakan seaborn
plt.figure(figsize=(12, 4))
sns.barplot(x='Genre Musik', y='Durasi Waktu (Menit per Hari)', hue='Usia', data=df)

#Menambahkan judul dan label
plt.title('Rata-rata Durasi Waktu Mendengarkan Musik')
plt.xlabel('Genre')
plt.ylabel('Durasi Waktu (Hitung pakai min)')

#Menampilkan plot
plt.show()
```

Setelah melakukan visualisasi, kita dapat melihat dengan mudah hasilnya, maka interpretasinya adalah :

*   Dapat dilihat disini bahwa, bahwa dari rentang 18-30 tahun, banyak anak muda yang menyukai genre pop. Namun dapat dilihat bahwa orang berumur 24-36 tahun juga menggemari genre jazz. Sedangnakan genre rock dapat dilihat dengan rentang 30-42 tahun. Terkait lagu klasik, tergolong pada golongan cenderung lebih tua.

*   Lalu, dapat dilihat dari durasi waktu mendengarkan musik bahwa kelompok usia 18-30 tahun memiliki durasi waktu yang tinggi namun hanya di genre pop saja. Selain itu, diparameter 42-48 tahun, cenderung memiliki durasi waktu yang tidak terlalu tingi dan lebih banyak mendengar lagu rock.

*   Genre classical dan rock lebih digemari oleh kelompok usia tua, maka kelompok usianya adalah 45 dan 50. Selain dari itu, dikalangan muda lebih suka menghabiskan waktu di genre rock
