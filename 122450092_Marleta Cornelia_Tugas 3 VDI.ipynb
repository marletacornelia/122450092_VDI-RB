**Visualisasi**

Sumber dataset : https://www.kaggle.com/datasets/meeratif/spotify-top-artists-by-monthly-listeners

Disini diambil beberapa sampel dan ingin membandingkan popularitas antara artis The Weeknd, Taylor Swift, Ed Sheeran, Dua Lipa, Bad Bunny,
               Rihanna, Drake, Justin Bieber, dan Billie Eilish.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#Mengambil beberapa sampel dan dibuat dataset
data = {
    'Artist': ['The Weeknd', 'Taylor Swift', 'Ed Sheeran', 'Dua Lipa', 'Bad Bunny',
               'Rihanna', 'Drake', 'Justin Bieber', 'Billie Eilish'],
    'Listeners': [107592328, 101003302, 76475126, 76421916, 76162057, 75784389,
                  75371611, 72623228, 71793820],
    'Daily Trend': [-138880, 889, -68137, -71356, -199052, 116405, -67013, 34540, -72877],
    'Peak': [1, 2, 2, 4, 3, 2, 6, 6, 8],
    'PkListeners': [113034886, 101003302, 87934910, 77778397, 83950570, 80958750,
                    76391086, 75467229, 72368528]
}

df = pd.DataFrame(data)

sns.set(style="whitegrid")

# Visualisasi 1: Bar Plot Jumlah Pendengar per Artis
plt.figure(figsize=(10, 6))
sns.barplot(x='Listeners', y='Artist', data=df, palette="Blues_d")
plt.title('Jumlah Pendengar per Artis')
plt.xlabel('Jumlah Pendengar')
plt.ylabel('Artis')
plt.show()

# Visualisasi 2: Scatter Plot Perubahan Harian Pendengar
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Daily Trend', y='Artist', size='Listeners', hue='Listeners', data=df, sizes=(100, 1000), palette="coolwarm", legend=False)
plt.title('Perubahan Harian Pendengar per Artis')
plt.xlabel('Perubahan Harian Pendengar')
plt.ylabel('Artis')
plt.show()
```

Setelah melakukan visualisasi, kita dapat melihat dengan mudah hasilnya, maka interpretasinya adalah :

*   Dapat dilihat disini bahwa, The Weeknd merupakan artis dengan jumlah pendengar harian terbanyak, namun terdapat penurunan tren di hsaat ini. Terlihat dari hasil -150000
*   Bad Bunny memiliki jumlah pendengar yang banyak dan tren hariannya mengalami penurunan yang lumayan jauh.

*   Rihanna dan Justin Bieber memiliki peningkatan di tren jumlah pendengar harian.
